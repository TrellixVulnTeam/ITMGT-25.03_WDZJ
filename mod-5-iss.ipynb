```python
import requests as req
import matplotlib.pyplot as plt
import pandas as pd
import json
import time

data = []

for x in range(0,100):
    res = req.get("http://api.open-notify.org/iss-now.json")
    text = res.text
    obj = json.loads(text)

    # add values in list
    # x = obj['iss_position']['latitude']
    # y = obj['iss_position']['longitude']
    
    data.append((obj["timestamp"], obj["iss_position"]["longitude"], obj["iss_position"]["latitude"]))
    time.sleep(10)
    
df = pd.DataFrame(data, columns = ["Timestamp", "Longitude", "Latitude"])
print(pd.DataFrame(df))
```

```python
df = pd.DataFrame(data, columns = ['Timestamp', 'Longitude','Latitude'])
ax1 = df.plot.scatter(x = "Latitude", y = "Longitude", c = "Blue", figsize = (25,25))
ax1.set_title("ISS Location")

plt.show()
```
